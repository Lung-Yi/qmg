```python
import qutip
import numpy as np

b = qutip.Bloch()
b.make_sphere()
# x = (qutip.basis(2, 0) + (1+0j)*qutip.basis(2, 1)).unit()
# y = (qutip.basis(2, 0) + (0+1j)*qutip.basis(2, 1)).unit()
# z = (qutip.basis(2, 0) + (0+0j)*qutip.basis(2, 1)).unit()

# b.add_states([x, y, z])
vec = [0,1,0]

b.add_vectors(vec, colors=["red"])

b.show()
b.render()
```

```python
xp = [np.cos(th) for th in np.linspace(0, 2*np.pi, 30)]
yp = np.zeros(30) 
zp = [np.sin(th) for th in np.linspace(0, 2*np.pi, 30)]
pnts = [xp, yp, zp]
b.add_points(pnts)

b.show()
```
