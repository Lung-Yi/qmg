```python
from pikachu.general import draw_smiles, highlight_substructure
from rdkit import Chem
draw_smiles("C[C@H](O)/N=C/[C@](C)(N)O")
```

```python
highlight_substructure("O1CC1", "CN(N)C1OC1N")
highlight_substructure("O1CC1", "NC12OC1n1on12")
highlight_substructure("O1CC1", "CNOCC1CO1")
highlight_substructure("O1CC1", "NN(O)C1(N)CO1")
highlight_substructure("O1CC1", "NNCCC1CO1")
highlight_substructure("C1OC1", "OC1=NC2(CO2)N1")
# highlight_substructure("O1CC1", "CN(N)C1OC1N", visualisation="svg", out_file="01.svg")
```

```python
smiles_dict = {'N=C=C(N)C(=O)CN': 1, 'NCC(=O)CCNNO': 1, 'CC(=O)/C=C\\N(C)NN': 1, 'CC(=O)CCN(N)N=N': 1, 'NCC(=O)[C@H]1CNC1=O': 1, 'N/N=C/CC(=O)CC=O': 1,}
for smiles in smiles_dict.keys():
    smiles = Chem.MolToSmiles(Chem.MolFromSmiles(smiles), isomericSmiles=False)
    highlight_substructure("CC(=O)C", smiles)
```
